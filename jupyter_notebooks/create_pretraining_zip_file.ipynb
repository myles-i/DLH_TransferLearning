```python
import zipfile
import os

# Directory containing the files (you can change this to your specific folder path)
directory = '/content/drive/MyDrive/DLHProject/data/icentia11k'

# Create a zip file with no compression
zip_filename = '/content/drive/MyDrive/DLHProject/data/icentia11k0to2047.zip'
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_STORED) as zipf:
    for i in range(2048):  # From 00000 to 02047
        file_number = f'{i:05}'  # Format numbers as 00000, 00001, ..., 02047
        lbls_file = f'{file_number}_batched_lbls.pkl.gz'
        batched_file = f'{file_number}_batched.pkl.gz'

        # Check if files exist and add them to the zip file
        if os.path.exists(os.path.join(directory, lbls_file)):
            zipf.write(os.path.join(directory, lbls_file), lbls_file)
        if os.path.exists(os.path.join(directory, batched_file)):
            zipf.write(os.path.join(directory, batched_file), batched_file)

print(f'Created zip file {zip_filename} without compression.')
```
